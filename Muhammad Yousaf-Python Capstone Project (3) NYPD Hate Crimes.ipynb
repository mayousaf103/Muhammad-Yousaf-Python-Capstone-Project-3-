3. EXAMINATION AND COMPARISON OF DATA

```python
#Creating a new dataframe for the Patrol Borough Manhattan South
Borough_One=['[PATROL BORO MAN SOUTH)']
```

```python
Borough_One=df[df['Patrol Borough Name'] == 'PATROL BORO MAN SOUTH']
```

```python
#Confirming the dataframe
Borough_One
```

```python
#Total of 414 rows and 11 columns
Borough_One.shape
```

```python
#Creating a second dataframe for the Patrol Borough Staten Island to compare to the first dataframe
Borough_Two=['[PATROL BORO STATEN ISLAND)']
```

```python
Borough_Two=df[df['Patrol Borough Name'] == 'PATROL BORO STATEN ISLAND']
```

```python
#Confirming the dataframe
Borough_Two
```

```python
#Total of 71 rows and 11 columns
Borough_Two.shape
```

```python
#Will merge both of the patrol boroughs to create a new dataframe
Merge=pd.concat([Borough_One, Borough_Two], ignore_index=True)
```

```python
#Confirming the merge dataframe
Merge
```

```python
#Total of 485 rows and 11 columns
Merge.shape
```

4. MEASURES OF CENTRAL TENDENCY

```python
#Aggregating Data/Grouping
Merge.groupby(['Record Create Date'])['Month Number'].count()
```

```python
#To display the mean value by complaint year number for both the counties
Merge.groupby(['County'])['Complaint Year Number'].mean()
```

```python
#To display the mean value by complaint ID based on the biased motives
Merge.groupby(['Bias Motive Description'])['Full Complaint ID'].mean()
```

```python
#To view the min value by counties for the precinct code
Merge.groupby(['County'])['Complaint Precinct Code'].min()
```

```python
#To view the max value by the categories of the law code for the precinct code as well
Merge.groupby(['Law Code Category Description'])['Complaint Precinct Code'].max()
```

```python
#To see the median value for the year the complaints were filed
Merge['Month Number'].median()
```

```python
#To see the mode for the dates for when the records were created  
Merge['Record Create Date'].mode()
```

```python
#To see if the pd code column is unique or not
Merge['PD Code Description'].is_unique
```

```python
#To see if the offense description column is unique or not
Merge['Offense Description'].is_unique
```

```python
#To display the count by the dates of the records for the county of new york for manhattan south 
Borough_One.groupby(['County']) ['Record Create Date'].count()
```

```python
#To see the median by the year in which the complaints were made according to the offenses committed for staten island
Borough_Two.groupby(['Offense Category']) ['Complaint Year Number'].median()
```
